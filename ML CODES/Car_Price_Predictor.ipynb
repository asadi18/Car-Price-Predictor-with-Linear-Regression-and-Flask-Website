# Preprocessing 
- Convert price , km_driven and year into int as they may have few issues. 
- Fill null values
- Keep first 3 words from name column to be clear 


```python
# at first working on year column 

df = df[df['year'].str.isnumeric()]
```

```python
df.info()
```

```python
df['year'] = df['year'].astype(int)
```

```python
df.info()
```

```python
df['Price'].unique()
```

```python
df = df[df['Price'] !="Ask For Price"]
```

```python
df['Price'] = df['Price'].str.replace(',','').astype(int)
```

```python
df.info()
```

```python
df['kms_driven'].unique()
```

```python
# Split and convert in int

df['kms_driven'] = df['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
df = df[df['kms_driven'].str.isnumeric()]
```

```python
df['kms_driven']= df['kms_driven'].astype(int)
```

```python
df.info()
```

```python
# Removing null values from fuel type
df = df[~df['fuel_type'].isna()]
```

```python
# Working on first column name making that short

df['name'] = df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df
```

```python
# From above we see the index has no proper values. so reseting old index
df = df.reset_index(drop=True)
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# as we seen a outlier here 85lakh car, only one piece we will remove it. and also see whichever is above 60lakh will be removed
df = df[df['Price']<6000000].reset_index(drop=True)
```

```python
df
```

```python
#  As we have cleaned the dataset we will save it into a csv file 
df.to_csv('cleaned_car_dataset.csv')
```

# Model building

```python
X = df.drop(columns='Price')
y = df['Price']
```

```python
y # as it is output other features are input
```

```python
# Importing train test and split 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
y_train  # Chceking all the values of X_train, X_test, y_train, y_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name', 'company', 'fuel_type']])
```

```python
col_trans = make_column_transformer((OneHotEncoder(categories = ohe.categories_), ['name', 'company', 'fuel_type']),remainder = 'passthrough')
```

```python
col_trans
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(col_trans, lr)
```

```python
pipe
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
# We need to improve r2 score and make the model more efficient 
scores = []
for i in range(1000):
    X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2, random_state = i)
    pipe = make_pipeline (col_trans,lr)
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    scores.append(r2_score(y_test, y_pred))
```

```python
# We need to improve r2 score and make the model more efficient 

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2, random_state = np.argmax(scores))
pipe = make_pipeline (col_trans,lr)
pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)
r2_score(y_test, y_pred)
```
